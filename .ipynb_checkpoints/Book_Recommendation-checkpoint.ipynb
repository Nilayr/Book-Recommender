
Analyze the distribution of book ratings, handle age outliers , handle missing age values in `users_df` by replacing values above 90 and below 5 with NaN and then imputing NaN with the median, and handle implicit ratings in `ratings_df` by removing them.



```python
import numpy as np

print("\nValue counts of 'Book-Rating' column in Ratings DataFrame:")
display(ratings_df['Book-Rating'].value_counts().sort_index())

# Handle missing age values in users_df
users_df['Age'] = users_df['Age'].apply(lambda x: np.nan if x > 90 or x < 5 else x)


users_df['Age'] = users_df['Age'].fillna(users_df['Age'].median())


print("\nMissing values in Users DataFrame after handling Age:")
display(users_df.isnull().sum())

# Handle implicit ratings in ratings_df
ratings_df = ratings_df[ratings_df['Book-Rating'] != 0]
print("\nRatings DataFrame after removing implicit ratings:")
display(ratings_df.head())
```

```python
import pandas as pd
from sklearn.model_selection import train_test_split
```

```python
ratings = pd.read_csv("Ratings.csv", encoding='latin1')
```

```python
ratings = ratings[ratings['Book-Rating'] > 0]
```

```python
train_df, test_df = train_test_split(ratings, test_size=0.2, random_state=42)
```

```python
print("Train size:", train_df.shape)
print("Test size:", test_df.shape)
```

```python
book_avg_rating = train_df.groupby('ISBN')['Book-Rating'].mean().to_dict()

def predict_rating(user_id, book_id):
    
    if book_id in book_avg_rating:
        return book_avg_rating[book_id]
    else:
        return train_df['Book-Rating'].mean()
```

```python
from sklearn.metrics import mean_squared_error, mean_absolute_error
import numpy as np

actual = []
predicted = []

for idx, row in test_df.iterrows():
    uid = row['User-ID']
    bid = row['ISBN']
    true_rating = row['Book-Rating']
    
    pred_rating = predict_rating(uid, bid)
    
    actual.append(true_rating)
    predicted.append(pred_rating)

# Compute RMSE and MAE
rmse = np.sqrt(mean_squared_error(actual, predicted))
mae = mean_absolute_error(actual, predicted)

print(f"RMSE: {rmse:.2f}")
print(f"MAE: {mae:.2f}")
```

```python
import matplotlib.pyplot as plt

metrics = ['RMSE', 'MAE']
values = [rmse, mae]

plt.figure(figsize=(6,4))
bars = plt.bar(metrics, values, color=['skyblue', 'orange'])

for bar in bars:
    yval = bar.get_height()
    plt.text(bar.get_x() + 0.25, yval + 0.01, f'{yval:.2f}')

plt.title('Evaluation Metrics for Rating Prediction')
plt.ylabel('Error')
plt.ylim(0, max(values) + 1)
plt.grid(axis='y', linestyle='--', alpha=0.7)
plt.show()
```

```python
import numpy as np

def precision_recall_at_k(user_actual, N=5):
    total_precision = []
    total_recall = []

    for user_id in user_actual:
        actual_books = user_actual[user_id]
        recommended_books = set(recommend_books(user_id, N))  # Your top-N recs

        true_positives = len(actual_books & recommended_books)
        precision = true_positives / N
        recall = true_positives / len(actual_books) if actual_books else 0

        total_precision.append(precision)
        total_recall.append(recall)

    avg_precision = np.mean(total_precision)
    avg_recall = np.mean(total_recall)

    return avg_precision, avg_recall
```

```python
from collections import defaultdict

# Define the rating threshold
relevant_threshold = 7

# Ground truth relevant books per user
user_actual = defaultdict(set)

# Loop through test set
for _, row in test_df.iterrows():
    if row['Book-Rating'] >= relevant_threshold:
        user_actual[row['User-ID']].add(row['ISBN'])
```

```python
# Top-N globally popular books from training data
top_books = (
    train_df.groupby('ISBN')['Book-Rating']
    .mean()
    .sort_values(ascending=False)
    .index
    .tolist()
)

def recommend_books(user_id, N=5):
    # Returns the same top-N books for all users (baseline)
    return top_books[:N]
```

```python
precision, recall = precision_recall_at_k(user_actual, N=5)
print(f"Precision@5: {precision:.2f}")
print(f"Recall@5: {recall:.2f}")
```

```python
# Filter users with at least 5 ratings
active_users = train_df['User-ID'].value_counts()
active_users = active_users[active_users >= 5].index

# Filter books with at least 10 ratings
popular_books = train_df['ISBN'].value_counts()
popular_books = popular_books[popular_books >= 10].index

filtered_train = train_df[
    train_df['User-ID'].isin(active_users) &
    train_df['ISBN'].isin(popular_books)
]
```

```python
user_item_matrix = filtered_train.pivot_table(index='User-ID', columns='ISBN', values='Book-Rating').fillna(0)
```

```python
from sklearn.metrics.pairwise import cosine_similarity

item_similarity = cosine_similarity(user_item_matrix.T)
item_similarity_df = pd.DataFrame(item_similarity,
                                   index=user_item_matrix.columns,
                                   columns=user_item_matrix.columns)
```

```python
def recommend_books(user_id, N=5):
    # If user not in matrix (cold start), fallback to global popular books
    if user_id not in user_item_matrix.index:
        return top_books[:N]

    # Get the user's rated books
    user_ratings = user_item_matrix.loc[user_id]
    rated_books = user_ratings[user_ratings > 0].index.tolist()

    scores = {}

    for book in rated_books:
        if book not in item_similarity_df:
            continue
        
        similar_books = item_similarity_df[book].drop(book)
        user_rating = user_ratings[book]
        
        for sim_book, similarity in similar_books.items():
            if sim_book in rated_books:
                continue  # skip books already rated
            scores.setdefault(sim_book, 0)
            scores[sim_book] += similarity * user_rating

    # Sort and return top-N recommendations
    recommended = sorted(scores.items(), key=lambda x: x[1], reverse=True)
    recommended_books = [isbn for isbn, _ in recommended[:N]]
    
    return recommended_books
```

```python
precision, recall = precision_recall_at_k(user_actual, N=5)
print(f"Precision@5: {precision:.2f}")
print(f"Recall@5: {recall:.2f}")
```

```python
import matplotlib.pyplot as plt

metrics = ['Precision@5', 'Recall@5']
values = [precision, recall]

plt.figure(figsize=(6,4))
bars = plt.bar(metrics, values, color=['mediumseagreen', 'coral'])

for bar in bars:
    yval = bar.get_height()
    plt.text(bar.get_x() + 0.15, yval + 0.01, f'{yval:.2f}')

plt.title('Top-N Recommendation Evaluation (Item-Based CF)')
plt.ylabel('Score')
plt.ylim(0, 1.0)
plt.grid(axis='y', linestyle='--', alpha=0.6)
plt.show()
```
